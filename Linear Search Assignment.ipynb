#Linear Search Implementation

```python
def linearSearch(arr, key):
    for x in arr:
        if x == key:
            return True
    return False
```

○ Count the number of key comparisons for various inputs and plot the graph.

○ For every input size n , run it with n+ 1 different keys - n successful plus 1 unsuccessful

○ Compute the minimum, maximum and average of the number of key comparisons for each input size.

○ Plot the graph for each case - best, worst and average number of comparisons.

○ n varies from 10 to 100 in steps of 5

```python
import random
import matplotlib.pyplot as plt

def linear_search(arr, key):
    comparisons = 0
    for ele in arr:
        comparisons += 1
        if ele == key:
          comparisons += 1
          return comparisons
    comparisons += 1
    return comparisons

input_sizes = list(range(10, 101, 5))
best_cases = []
worst_cases = []
average_cases = []


for n in input_sizes:
    arr = list(range(1, n + 1))

    best_cases.append(linear_search(arr, arr[0]))

    worst_cases.append(linear_search(arr, n + 1))

    total_comparisons = 0
    for _ in range(n):
        key = random.choice(arr)
        total_comparisons += linear_search(arr, key)
    total_comparisons += linear_search(arr, n + 1)
    average_cases.append(total_comparisons / (n + 1))

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, best_cases, label="Best Case", marker='o')
plt.plot(input_sizes, worst_cases, label="Worst Case", marker='s')
plt.plot(input_sizes, average_cases, label="Average Case", marker='^')

plt.title("Linear Search Comparisons")
plt.xlabel("Input Size")
plt.ylabel("Number of Comparisons")
plt.legend()
plt.grid()
plt.show()
```
