```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linear_search(arr, key):
    """Performs linear search and returns the number of comparisons."""
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return comparisons
    return comparisons
```

```python
# Parameters
input_sizes = range(10, 101, 5)  # n varies from 10 to 100 in steps of 5
best_case = []
worst_case = []
average_case = []
```

```python
for n in input_sizes:
    arr = list(range(1, n + 1))  # Create an array of size n (1 to n)
    total_comparisons = 0

    # Best case: Key is the first element
    key = arr[0]
    best_case.append(linear_search(arr, key))

    # Worst case: Key is not in the array (unsuccessful search)
    key = n + 1
    worst_case.append(linear_search(arr, key))

    # Average case: Run for all n elements (successful) and 1 unsuccessful
    for key in arr:
        total_comparisons += linear_search(arr, key)
    total_comparisons += linear_search(arr, n + 1)  # Add unsuccessful search

    avg_comparisons = total_comparisons / (n + 1)
    average_case.append(avg_comparisons)
```

```python
# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(input_sizes, best_case, label="Best Case", marker="o")
plt.plot(input_sizes, worst_case, label="Worst Case", marker="o")
plt.plot(input_sizes, average_case, label="Average Case", marker="o")

plt.title("Linear Search: Comparisons vs Input Size")
plt.xlabel("Input Size (n)")
plt.ylabel("Number of Comparisons")
plt.legend()
plt.grid()
plt.show()
```
